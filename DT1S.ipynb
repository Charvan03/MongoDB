```python
class node:
  def __init__(self,data): 
    self.left = None 
    self.data = data 
    self.right = None 
  def displayTree(self): 
    if self.left:
      self.left.displayTree()
    print(self.data)
    if self.right:
      self.right.displayTree()
    
  def insertTree(self,data):
    if self.data:
      if data < self.data:
        if self.left is None:
          self.left = node(data)
        else:
          self.left.insertTree(data)
      elif data > self.data:
        if self.right is None:
          self.right = node(data)
        else:
          self.right.insertTree(data)
      else:
        self.data = data
    
```
